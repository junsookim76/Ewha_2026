## Part 3. 분자 Descriptor 계산 (20분)

> 📚 **복습**: RDKit Descriptors, Numpy 배열 연산, Pandas DataFrame 다루기

### 배경 지식: Molecular Descriptors

분자의 물리화학적 성질을 숫자로 나타낸 것을 **descriptor**라고 합니다.

| Descriptor | 설명 | 함수 |
|------------|------|------|
| MW | 분자량 (Molecular Weight) | `Descriptors.MolWt()` |
| LogP | 지용성 (octanol-water 분배계수) | `Descriptors.MolLogP()` |
| HBD | 수소결합 donor 수 | `Descriptors.NumHDonors()` |
| HBA | 수소결합 acceptor 수 | `Descriptors.NumHAcceptors()` |
| TPSA | 극성 표면적 | `Descriptors.TPSA()` |
| RotBonds | 회전 가능한 결합 수 | `Descriptors.NumRotatableBonds()` |

### 3-1. Descriptor 계산 함수 만들기

> **할 일**: 빈칸을 채워 descriptor를 계산하는 함수를 완성하세요.

```python
def calculate_descriptors(smiles):
    """
    SMILES 문자열을 입력받아 분자 descriptor를 계산하는 함수

    Parameters:
        smiles (str): 분자의 SMILES 문자열

    Returns:
        dict: descriptor 이름과 값의 딕셔너리
    """
    mol = Chem.MolFromSmiles(smiles)

    if mol is None:
        return None

    descriptors = {
        'MW': Descriptors.MolWt(mol),                    # 분자량
        'LogP': Descriptors.______(mol),                  # 힌트: MolLogP
        'HBD': Descriptors.NumHDonors(mol),               # 수소결합 donor
        'HBA': Descriptors.______(mol),                   # 힌트: NumHAcceptors
        'TPSA': Descriptors.TPSA(mol),                    # 극성 표면적
        'RotBonds': Descriptors.NumRotatableBonds(mol),   # 회전 가능 결합
        'NumAtoms': mol.GetNumAtoms(),                    # 원자 수
        'NumRings': rdMolDescriptors.CalcNumRings(mol)    # 고리 수
    }

    return descriptors

# 테스트: Aspirin의 descriptor 계산
test_result = calculate_descriptors('CC(=O)Oc1ccccc1C(=O)O')
print("Aspirin descriptors:")
for key, value in test_result.items():
    print(f"  {key}: {value:.2f}")
```

### 3-2. 전체 분자에 대해 Descriptor 계산

> **할 일**: 모든 분자의 descriptor를 계산하여 DataFrame에 추가하세요.

```python
# 모든 분자의 descriptor를 계산합니다
results = []
for smiles in df['SMILES']:
    desc = calculate_descriptors(smiles)
    results.append(desc)

# descriptor 결과를 DataFrame으로 변환
desc_df = pd.DataFrame(results)

# 원래 데이터와 합치기
df_full = pd.concat([df, desc_df], axis=1)

# 결과 확인
df_full.head(10)
```

## Part 5. Lipinski's Rule of Five 적용 (20분)

> 📚 **복습**: IF 조건문, 함수 정의, Pandas 필터링, Matplotlib

### 배경 지식: Lipinski's Rule of Five

**"먹는 약이 될 수 있는가?"**를 판단하는 경험적 규칙입니다.

아래 **4가지 조건을 모두 만족**하면 경구 투여(oral drug)로 적합할 가능성이 높습니다:

| 규칙 | 조건 |
|------|------|
| Rule 1 | 분자량 (MW) ≤ **500** Da |
| Rule 2 | LogP ≤ **5** |
| Rule 3 | 수소결합 Donor (HBD) ≤ **5** |
| Rule 4 | 수소결합 Acceptor (HBA) ≤ **10** |

### 5-1. Lipinski 규칙 판별 함수 만들기

> **할 일**: 빈칸을 채워 Lipinski 규칙을 판별하는 함수를 완성하세요.

```python
def check_lipinski(row):
    """
    DataFrame의 한 행을 입력받아 Lipinski Rule of Five를 검사하는 함수

    Parameters:
        row: DataFrame의 한 행 (MW, LogP, HBD, HBA 열 포함)

    Returns:
        int: 위반 규칙 수 (0이면 모든 규칙 통과)
    """
    violations = 0

    if row['MW'] > ______:       # Rule 1: MW 기준은?
        violations += 1
    if row['LogP'] > ______:     # Rule 2: LogP 기준은?
        violations += 1
    if row['HBD'] > ______:      # Rule 3: HBD 기준은?
        violations += 1
    if row['HBA'] > ______:      # Rule 4: HBA 기준은?
        violations += 1

    return violations

# 모든 분자에 Lipinski 규칙 적용
df_full['Lipinski_Violations'] = df_full.apply(check_lipinski, axis=1)

# 결과 확인
df_full[['Name', 'Category', 'MW', 'LogP', 'HBD', 'HBA', 'Lipinski_Violations']].head(10)
```

### 5-2. Drug-likeness 판별

> **할 일**: Lipinski 규칙을 통과한 약물과 통과하지 못한 약물을 분류하세요.

```python
# Drug-like 여부 판별 (위반 0개 = Drug-like)
df_full['Drug_Like'] = df_full['Lipinski_Violations'] == ______  # 빈칸: 위반 0개일 때 True

# Drug-like vs Non-Drug-like 개수
drug_like_count = df_full['Drug_Like'].sum()
non_drug_like_count = len(df_full) - drug_like_count

print(f"Drug-like 분자: {drug_like_count}개")
print(f"Non-Drug-like 분자: {non_drug_like_count}개")
print(f"Drug-like 비율: {drug_like_count/len(df_full)*100:.1f}%")
```

```python
# Lipinski 규칙을 통과하지 못한 분자들 확인
non_drug_like = df_full[df_full['Drug_Like'] == False]
print("\n=== Non-Drug-like 분자 ===")
print(non_drug_like[['Name', 'Category', 'MW', 'LogP', 'HBD', 'HBA', 'Lipinski_Violations']])
```

### 5-3. Drug-likeness 시각화

> **할 일**: Drug-like 여부에 따라 분자들을 시각화해 보세요.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# (a) MW vs LogP scatter plot (Drug-like 여부로 색 구분)
drug_like = df_full[df_full['Drug_Like'] == True]
non_drug = df_full[df_full['Drug_Like'] == False]

axes[0].scatter(drug_like['MW'], drug_like['LogP'],
                color='steelblue', label='Drug-like', s=80, alpha=0.7, edgecolors='black')
axes[0].scatter(non_drug['MW'], non_drug['LogP'],
                color='crimson', label='Non-Drug-like', s=80, alpha=0.7, edgecolors='black', marker='x')

# Lipinski 기준선 그리기
axes[0].axvline(x=500, color='gray', linestyle='--', linewidth=1, alpha=0.7)
axes[0].axhline(y=5, color='gray', linestyle='--', linewidth=1, alpha=0.7)

axes[0].set_xlabel('Molecular Weight (Da)', fontsize=13)
axes[0].set_ylabel('LogP', fontsize=13)
axes[0].set_title('MW vs LogP (Lipinski Filter)', fontsize=14)
axes[0].legend(fontsize=11)

# (b) 위반 개수별 분포 (Pie Chart)
violation_counts = df_full['Lipinski_Violations'].value_counts().sort_index()
colors_pie = ['steelblue', 'gold', 'orange', 'crimson', 'darkred']

axes[1].pie(violation_counts.values,
            labels=[f'{v} violations' for v in violation_counts.index],
            colors=colors_pie[:len(violation_counts)],
            autopct='%1.1f%%', startangle=90, textprops={'fontsize': 12})
axes[1].set_title('Lipinski Violations Distribution', fontsize=14)

plt.tight_layout()
plt.show()
```

### 5-4. Non-Drug-like 분자 구조 확인

> **할 일**: Lipinski 규칙을 통과하지 못한 분자들의 구조를 시각화하여 **왜 규칙을 위반했는지** 생각해 보세요.

```python
# Non-Drug-like 분자들의 구조 시각화
non_drug_smiles = non_drug['SMILES'].values
non_drug_names = non_drug.apply(
    lambda x: f"{x['Name']}\n(MW={x['MW']:.0f}, LogP={x['LogP']:.1f})", axis=1
).values.tolist()

non_drug_mols = [Chem.MolFromSmiles(smi) for smi in non_drug_smiles]

# None이 아닌 분자만 필터링
valid = [(m, n) for m, n in zip(non_drug_mols, non_drug_names) if m is not None]
if valid:
    mols_v, names_v = zip(*valid)
    Draw.MolsToGridImage(list(mols_v), molsPerRow=3, subImgSize=(350, 350), legends=list(names_v))
else:
    print("Non-Drug-like 분자가 없습니다!")
```
